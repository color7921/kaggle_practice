```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import chromedriver_autoinstaller
import subprocess
import shutil
from datetime import datetime

import time
import requests
import random
import numpy as np
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# try:
#     shutil.rmtree(r"c:\chrometemp")  #쿠키 / 캐쉬파일 삭제
# except FileNotFoundError:
#     pass
    
subprocess.Popen(r'C:\Program Files\Google\Chrome\Application\chrome.exe --remote-debugging-port=9222 --user-data-dir="C:\chrometemp"') # 디버거 크롬 구동

option = Options()
option.add_experimental_option("debuggerAddress", "127.0.0.1:9222")

driver = webdriver.Chrome(options=option)

driver.implicitly_wait(1)

# -------/ url
url_list = []

# -------/ 
details_image = []
img_src_list = []
cate_name = []
split_cate = []

# -------/ 상품 이름
product_name_text = []
product_name = []

# -------/ 가격
after_price = []
origin_price = []

# -------/ 카테고리
cate1 = []
cate2 = []
cate3 = []
cate4 = []
cate5 = []

# -------/ 상품 설명
coupang_detail1 = []
coupang_detail2 = []
detail_1 = []
detail_2 = []

# -------/ 상품 이미지
details_image = []
get_image = []
thumnail_1 = []

# -------/ 브랜드
brand_name2 = []

# -------/ 사이트
site_name = []

count = 0

for index, row in df.iloc[20:100].iterrows():
    i = row['상품명']
    
    formatted_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    count += 1
    print(f'{count}. {i} : {formatted_time}')

    try:
        driver.get(f"https://www.google.com/search?q={i}")
        # 검색 결과에서 링크 찾기
        scroll = driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(1)
        links = driver.find_elements(By.TAG_NAME, "a")
        # driver.implicitly_wait(10)
    except:
        print("ssg_error1")
    # 모든 텍스트 출력 및 쿠팡 링크 클릭

    text = []
    b = []
    a = []
    c = []
    for link in links:
        text = link.text
        if "www.coupang.com › product" in text:
            b = link.get_attribute("href")
            a = "쿠팡"

        elif "www.ssg.com › item" in text or "m.ssg.com › item" in text or "emart.ssg.com › item" in text:
            b = link.get_attribute("href")
            a = "이마트"
            
        elif "prod.danawa.com › info" in text:
            b = link.get_attribute("href") 
            a = "다나와"
            
        c.append(text)

    
    # >>>> 쿠팡
    
    if '쿠팡' in a:
        driver.get(b)
        time.sleep(1)
        search = None
        try:
            search = driver.find_element(By.XPATH, '//*[@id="searchOptionForm"]/div[2]/div[2]/div[1]/p')
        except:
            pass
        if not search: 
            
            site_name.append("쿠팡")
    
            try:
                brand_ = driver.find_element(By.XPATH, '//*[@id="contents"]/div[2]/div[1]/div[3]/a').text
                brand_name2.append(brand_)
            except:
                brand_name2.append("none")
            # ----------------------------------------- url -------------------------------------
            site_url = driver.current_url
            url_list.append(site_url)
        
            # ----------------------------------------- category -------------------------------------
    
            try:
                cate1_1 = driver.find_element(By.XPATH, '//*[@id="breadcrumb"]/li[2]/a').text
                cate1.append(cate1_1)
            except:
                cate1.append("none")
            try:
                cate2_1 = driver.find_element(By.XPATH, '//*[@id="breadcrumb"]/li[3]/a').text
                cate2.append(cate2_1)
            except:
                cate2.append("none")
            try:
                cate3_1 = driver.find_element(By.XPATH, '//*[@id="breadcrumb"]/li[4]/a').text
                cate3.append(cate3_1)
            except:
                cate3.append("none")
            try:
                cate4_1 = driver.find_element(By.XPATH, '//*[@id="breadcrumb"]/li[5]/a').text
                cate4.append(cate4_1)
            except:
                cate4.append("none")
            try:
                cate5_1 = driver.find_element(By.XPATH, '//*[@id="breadcrumb"]/li[6]/a').text
                cate5.append(cate5_1)
            except:
                cate5.append("none")
            
            # ----------------------------------------- product_name -------------------------------------
            try:
                product_name_text = driver.find_element(By.XPATH, '//*[@id="contents"]/div[2]/div[1]/div[3]/div[3]/h2').text
                product_name.append(f'{i}|{product_name_text}')
            except:
                product_name.append("none")
            # print(f'쿠팡{product_name}')
        
            # ----------------------------------------- 상품 설명1 -------------------------------------
            try:
                coupang_detail1 = driver.find_element(By.CLASS_NAME, 'prod-description').text
                coupang_detail2 = driver.find_element(By.CLASS_NAME, 'product-item__table').text
            
                detail_1.append(coupang_detail1)
                detail_2.append(coupang_detail2)
            except:
                detail_1.append("none")
                detail_2.append("none")
            # ----------------------------------------- 상품 가격 -------------------------------------
            try:
                coupang_2 = driver.find_element(By.XPATH, '//*[@id="contents"]/div[2]/div[1]/div[3]/div[5]/div[1]/div/div[1]/span[2]').text
                origin_price.append(coupang_2)
            except:
                origin_price.append("none")
            try:
                coupang_3 = driver.find_element(By.XPATH, '//*[@id="contents"]/div[2]/div[1]/div[3]/div[5]/div[1]/div/div[2]/span/strong').text
                after_price.append(coupang_3)
            except:
                after_price.append("none")
 
            # ----------------------------------------- 상품 썸네일 이미지 -------------------------------------
            try:
                thumnail = driver.find_element(By.CLASS_NAME, 'prod-image__detail').get_attribute("data-zoom-image-url")
                thumnail_1.append(f'{i}|{thumnail}')
            except:
                pass
            # ----------------------------------------- 상품 상세 이미지 -------------------------------------
            try:
                elements1 = driver.find_elements(By.CSS_SELECTOR, '.vendor-item [src]')
                elements_img = []
                # src 출력
                # elements1 = []
                for element1 in elements1:
                    try:
                        get_image.append(element1.get_attribute('src'))
                    except:
                        get_image.append("none")
                details_image.append(f"{i}|{'|'.join(get_image)}")
            except:
                pass
                
                
        # >>>> 이마트

    elif "이마트" in a:  
        driver.get(b)
        # link.click()
        time.sleep(1)
        site_name.append("이마트")
        # ----------------------------------------- url -------------------------------------
        site_url = driver.current_url
        url_list.append(site_url)

        # ----------------------------------------- product_name -------------------------------------
        try:
            d = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[1]/div[2]/div[2]/h2/span/span').text
            product_name.append(f"{i}|{d}")
        except:
            product_name.append("none")
        # print(f'이마트{product_name}')
        # ----------------------------------------- 브랜드 -------------------------------------
        try:
            bn2 = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[1]/div[2]/div[2]/h2/span/div/div/a').text
            brand_name2.append(bn2)
        except:
            brand_name2.append("none")

        # ----------------------------------------- 이미지 -------------------------------------
        img_urls = []
        for j in range(1, 7):  # 이미지는 최대 6개까지 가져옴
            try:
                img = driver.find_element(By.XPATH, f'//*[@id="content"]/div[2]/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/ul/li[{j}]/a/img')
                img_urls.append(img.get_attribute('src'))
            except:
                pass
        try:
            img = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[1]/div[1]/div[1]/div[2]/div/div/ul/li/a/img')
            img_urls.append(img.get_attribute('src'))
        except:
            pass
        
        thumnail_1.append(f"{i}|{'|'.join(img_urls)}")

        try:
            e_2 = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[1]/div[2]/div[4]/div[1]/div/span[2]/em').text
            origin_price.append(e_2)
        except:
            origin_price.append(f'none')
        try:
            e_3 = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[1]/div[2]/div[4]/div[1]/div/span[1]/em').text
            after_price.append(e_3)
        except:
            after_price.append("none")

        # ----------------------------------------- 카테고리 -------------------------------------
        try:
            cate1.append(driver.find_element(By.XPATH, '//*[@id="location"]/div[2]/a').text)
        except:
            cate1.append("none")
        try:
            cate2.append(driver.find_element(By.XPATH, '//*[@id="location"]/div[3]/a').text)
        except:
            cate2.append("none")
        try:
            cate3.append(driver.find_element(By.XPATH, '//*[@id="location"]/div[4]/a').text)
        except:
            cate3.append("none")

        cate4.append("none")
        cate5.append("none")
        img_src_list = []
        try:
            # iframe ID : _ifr_html
            iframe_element = driver.find_element(By.XPATH, '//*[@id="_ifr_html"]')
            
            # iframe으로 전환
            driver.switch_to.frame(iframe_element)
    
            # div id가 "descContents"인 태그 내의 이미지 src 추출
            desc_contents_div = driver.find_element(By.ID, "descContents")
            img_elements = desc_contents_div.find_elements(By.TAG_NAME, "img")
            
            for img in img_elements:
                src = img.get_attribute("src")
                img_src_list.append(src)
            
            # 원래의 상위 프레임으로 전환
            driver.switch_to.default_content()
        except:
            pass
        try:
            details_image.append(f"{i}|{'|'.join(img_src_list)}")
        except:
            details_image.append(f"{i}|none")

        try:
            table = driver.find_element(By.CLASS_NAME, 'cdtl_tbl')
            tbody_elements = table.find_elements(By.TAG_NAME, 'tbody')
            
            for tbody in tbody_elements:
                detail_2.append(tbody.text.replace('\n', '|'))
        except:
            detail_2.append("none")
        detail_1.append("none")

    
        # >>>> 다나와
    
    elif "다나와" in a:  
        driver.get(b)
        # link.click()
        time.sleep(1)
        site_name.append("다나와")
        # ----------------------------------------- url -------------------------------------
        site_url = driver.current_url
        url_list.append(site_url)

        # ----------------------------------------- 제품명 -------------------------------------
        try:
            a_1 = driver.find_element(By.XPATH, '//*[@id="blog_content"]/div[2]/div[1]/h3/span').text
            product_name.append(f"{i}|{a_1}")
        except:
            product_name.append("none")
        # ----------------------------------------- 카테고리 -------------------------------------
        try:
            cate_1 = driver.find_element(By.XPATH, '//*[@id="sectionCategory"]/button').text
            cate1.append(cate_1)
        except:
            cate1.append("none")
        try:
            cate_2 = driver.find_element(By.XPATH, '//*[@id="1DepthCategory"]/button').text
            cate2.append(cate_2)
        except:
            cate2.append("none")
        try:
            cate_3 = driver.find_element(By.XPATH, '//*[@id="2DepthCategory"]/button').text
            cate3.append(cate_3)
        except:
            cate3.append("none")
        cate4.append("none")
        cate5.append("none")

        # ----------------------------------------- 상세설명 -------------------------------------
        try:
            spec_list = driver.find_element(By.XPATH, '//*[@id="blog_content"]/div[2]/div[1]/div/div[1]/div/dl/dd/div/div').text
            detail_2.append(spec_list)
        except:
            detail_2.append("none")
        detail_1.append("none")

        # ----------------------------------------- 가격 -------------------------------------
        try:
            price = driver.find_element(By.XPATH, '//*[@id="blog_content"]/div[2]/div[2]/div[1]/div[2]/div[1]/div/span[2]/a/em').text
            after_price.append(price)
        except:
            after_price.append("none")
        origin_price.append("none")

        # ----------------------------------------- 이미지 -------------------------------------
        try:
            img = driver.find_element(By.XPATH, '//*[@id="baseImage"]').get_attribute('src')
            thumnail_1.append(f"{i}|{img}")
        except:
            thumnail_1.append(f"{i}|none")
        details_image.append(f"{i}|none")
        brand_name2.append("none")

    time.sleep(random.uniform(2,5))
```
